```python
## Clean and regroup all gold players
g_player = player_details[(player_details['Overall'] >= 80)]
g_id = g_player['ID']
g_id.values[0]
columns = ['ID','Player Name','rating','Date','potential','height','weight','birthday','fifa id','age']
gold_player = pd.DataFrame(columns = columns)

for _id in g_id.values:
    players = player_df[player_df['fifa id'] == _id]
    result = gold_player.append(players)
    gold_player = result
gold_player.fillna(0, inplace=True)
gold_player = gold_player[gold_player["rating"] != 0].sort_values(["Player Name"], ascending=[0])
gold_player
```

* From player_df dataframe i regroup all gold player in a single table.

```python


# getting a random player by age
def get_random_sample_player_rate(age):
    player = gold_player[(gold_player["age"] == age)].sample(1)
    return player['rating'].values[0]

#calculate sample mean
def get_player_sample_mean(age, n = 50):
  
    player_set = np.empty((0,n))
    for _ in range(0, n):
       
        single_rate = get_random_sample_player_rate(age)
        player_set = np.append(player_set, single_rate)
        
    
    player_mean = np.sum(player_set)/n
    return player_mean

#calculate mean distribution
def player_rating_by_age_distribution(age, size = 1000):
    
    mean_set = np.empty((0,size))
    for _ in range(0, size):
        n = random.randint(30,100)
        player_mean = get_player_sample_mean(age, n)
        mean_set = np.append(mean_set, player_mean)
    return mean_set 
    
```

```python
# tabel max, mean, and min by ages
def mean_by_age(age):
    mean = gold_player[gold_player['age'] == age].mean()
    return mean['rating']
def quatile_max_by_age(age):
    _max = gold_player[gold_player['age'] == age].quantile(.75)
    return _max['rating']
def quatile_min_by_age(age):
     _min = gold_player[gold_player['age'] == age].quantile(.25)
     
     return _min['rating']
```

```python
columns = ['age','poor','low','average', 'good','excellent']
player_evolution = pd.DataFrame(columns=columns)

for age in range(16,49):
    low = quatile_min_by_age(age)
    good = quatile_max_by_age(age)
    avg = mean_by_age(age)
    worst = gold_player[gold_player['age'] == age].min()['rating']
    excellent = gold_player[gold_player['age'] == age].max()['rating']
    bad = gold_player[gold_player['age'] == age].min()['rating']
    if type(bad) is int: 
        player_evolution.loc[age] = ([int(age),int(bad),int(low),int(avg),int(good),int(excellent)])
```

```python
shart = player_evolution[player_evolution['age'] > 15]
ax = player_evolution[['poor','low','average', 'good','excellent']].plot(kind='line', title = "PLayer Evolution", figsize=(20, 7), legend=True, fontsize=15)
ax.set_xlabel("Age", fontsize=12)
ax.set_ylabel("Rating", fontsize=20)
sns.palplot(sns.color_palette("RdBu_r", 1))
plt.show()
```

```python
def asign_prog(age,rate):
   
    pl = player_evolution[player_evolution['age'] == age]
    if rate == pl['poor'].values[0]:
        return 'bad progression'
    elif rate in range( pl['poor'].values[0] ,  pl['low'].values[0]):
        return 'low progression'
    elif rate in range(pl['low'].values[0] ,  pl['average'].values[0]):
        return 'average progression'
    elif rate in range (pl['average'].values[0] , pl['good'].values[0]):
        return 'good progression'
    elif rate in range (pl['good'].values[0] , pl['excellent'].values[0] + 1):
        return 'excellent progression'
#     elif rate >= pl['excellent'].values[0]:
#         return 'excellent progression'
    else: return 'none'
    
progress_dict = {'none':0,"bad progression":1,'poor progression':2,"low progression":3,'average progression':4,'good progression':5,'excellent progression':6}
```

```python

_id = final_dataset['ID'].values

for index, row in final_dataset.iterrows():
    age = row['Age']
    if age <= 38:
        rate = row['Overall']
        asg = asign_prog(age,rate)
        final_dataset.loc[index, "progression status"] = asg
        final_dataset.loc[index, "progression value"] = progress_dict[asg]
    else:
        final_dataset.loc[index, "progression status"] = 'none'
        final_dataset.loc[index, "progression value"] = 0.0
  
```
